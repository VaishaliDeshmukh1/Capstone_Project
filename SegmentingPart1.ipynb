#### Import Required libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import numpy as np
```

#### Access the url & print the required table using urllib.request url (open function) or by using request.get function

```python
#from urllib.request import urlopen
#url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
#source= urlopen(url) #
#soup=BeautifulSoup(Source,"lxml")
#print(soup.prettify())
```

```python
source=requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
soup=BeautifulSoup(source,"lxml") # lxml is a parser for html doc
#print(soup.prettify()) # prettify is used to format the unicode string # Remove hash symbol to print the data
```

#### Inpect the weblink (right click on web page and click on inspect) & find the Tags from where table is created

```python
table= soup.find("table",class_="wikitable sortable").tbody # find will help to find all the tags of our table
#print(table.prettify()) # Remove hash symbol to print the data 
```

#### Check rows and columns from our table which we have create in previous step

```python
rows=table.find_all("tr")
# "tr" is tag of html for defining table rows

column_names= [v.text.replace("\n","") for v in rows[0].find_all("th")]
# "th" is tag of html for defining table heading or columns
# v.text.replace("\n","") --> this will replace the \n with a blank 

column_names
```

#### Create a dataframe with columns

```python
df = pd.DataFrame(columns=column_names) 
df
```

#### Create a for loop for itterating through the table to find the table data(td)

```python
for i in range (1,len(rows)):
    tds=[v.text.replace("\n","") for v in rows[i].find_all("td")] # td tag in html used to define the data in table
    
    df=df.append(pd.Series(tds,index=column_names),ignore_index=True) # append the table data in our df created in previous step
    print (df)
```

#### In Borough column, replace the value of Not assigned with the numpy nan(not a number) in order to drop it in next step

```python
df["Borough"].replace("Not assigned",np.nan,inplace=True) # np.nan is numpy default function to assign NaN 
```

```python
df1=df.dropna() # this will drop the NaN values from the Table
df1.head(10)
```

```python
df1.reset_index(drop=True,inplace=True) # this will reset the index from 0
df1.head(10)
```

#### Use Groupby to combine the rows of same Postcode values and to seperate their value by comma

```python
df1 = df1.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()  
df1.columns = ['Postcode', 'Borough', 'Neighbourhood']
df1.head(10)
```

#### Assign values of Borough where the value of neighbourhood is Not assigned

```python
df1.loc[df1['Neighbourhood']=="Not assigned",'Neighbourhood']=df1.loc[df1['Neighbourhood']=="Not assigned",'Borough']
df1.head(10)
```
